```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import sys
import operator
```

```python
def getTopFive(dic):
    return  sorted(dic.items(), key=operator.itemgetter(1),reverse = True )[:5]
 
def printTopFive(dic):
    topFive = getTopFive(dic)
    for index in range(5):        
        print(str((index+1))+" "+str(topFive[index][0])+" "+"%.2f" % (topFive[index][1]))
```

```python
#Q2.a(algorithm.py)
import algorithm
#Q2.b

graph = nx.florentine_families_graph()
degResult = algorithm.degreeCentrality(graph)
closResult = algorithm.closenessCentrality(graph)
betResult = algorithm.betweennessCentrality(graph)

print('\n TOP 5 \n Degree Centrality:')
printTopFive(degResult)
print('\n TOP 5 \n Closeness Centrality:')
printTopFive(closResult)
print('\n TOP 5 \n Betweenness Centrality:')
printTopFive(betResult)
```
